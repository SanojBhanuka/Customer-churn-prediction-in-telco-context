```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Define continuous variables
continuos_var = df[['tenure', 'MonthlyCharges', 'TotalCharges']]

# Standardize the data
scaler = StandardScaler()
std_continuos_var = scaler.fit_transform(continuos_var)
std_continuos_var = pd.DataFrame(std_continuos_var, columns=continuos_var.columns)
std_continuos_var.index = continuos_var.index

# Plot before and after standardization
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))

# Plot before standardization
for i, column in enumerate(continuos_var.columns):
    sns.histplot(continuos_var[column], kde=True, ax=axes[0, i], color='blue', label='Before Standardization')
    axes[0, i].set_title(f'Distribution of {column} (Before Standardization)')
    axes[0, i].set_xlabel(column)
    axes[0, i].set_ylabel('Density')
    axes[0, i].legend()

# Plot after standardization
for i, column in enumerate(std_continuos_var.columns):
    sns.histplot(std_continuos_var[column], kde=True, ax=axes[1, i], color='orange', label='After Standardization')
    # Plot standard normal distribution curve
    x = np.linspace(-4, 4, 1000)
    y = 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x ** 2)
    axes[1, i].plot(x, y, color='green', linestyle='--', label='Standard Normal Distribution')

    axes[1, i].set_title(f'Distribution of {column} (After Standardization)')
    axes[1, i].set_xlabel(column)
    axes[1, i].set_ylabel('Density')
    axes[1, i].legend()

plt.tight_layout()
plt.show()
```

## **Spliting the dataset**

```python

x = df_cleaned.drop(['Churn'], axis=1)
y = df_cleaned['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
y_train.value_counts()
```

##**Handle class imbalance (SMOTE)**

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE(random_state=0)
x_train_smote, y_train_smote = oversample.fit_resample(x_train,y_train)
#Display the class distribution after SMOTE
print(y_train_smote.value_counts())
```

```python
import matplotlib.pyplot as plt

# Calculate class distribution before SMOTE
class_distribution_before = y_train.value_counts()

# Calculate class distribution after SMOTE
class_distribution_after = y_train_smote.value_counts()

# Create subplots for the two bar graphs
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot class distribution before SMOTE
axes[0].bar(class_distribution_before.index, class_distribution_before.values)
axes[0].set_title('Class Distribution Before SMOTE')
axes[0].set_xlabel('Class')
axes[0].set_ylabel('Frequency')

# Plot class distribution after SMOTE
axes[1].bar(class_distribution_after.index, class_distribution_after.values)
axes[1].set_title('Class Distribution After SMOTE')
axes[1].set_xlabel('Class')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## **Model Training and Evaluation (Logistic Regression)**

```python
# Model

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_smote,y_train_smote)
```

```python
# Training set

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,f1_score, precision_score, recall_score
y_train_pred = model.predict(x_train_smote)
train_acc = accuracy_score(y_train_smote, y_train_pred)
train_err = 1 - train_acc
train_f1 = f1_score(y_train_smote, y_train_pred, average='weighted')
train_cm = confusion_matrix(y_train_smote, y_train_pred)
train_sensitivity = train_cm[1,1]/(train_cm[1,1]+train_cm[1,0])
train_specificity = train_cm[0,0]/(train_cm[0,0]+train_cm[0,1])
```

```python

print(f'Training Accuracy: {train_acc:.4f}')
print(f'Training Error Rate: {train_err:.4f}')
print(f'Training F1 Score: {train_f1:.4f}')
print(f'Training Confusion Matrix:\n{train_cm}')
print("Training sensitivity: ", train_sensitivity)
print("Training specificity: ", train_specificity)
```

```python
# Testing set
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, precision_score, recall_score
y_test_pred = model.predict(x_test)
test_acc = accuracy_score(y_test, y_test_pred)
test_err = 1 - test_acc
test_f1 = f1_score(y_test, y_test_pred, average='weighted')
test_cm = confusion_matrix(y_test, y_test_pred)
test_sensitivity = test_cm[1, 1] / (test_cm[1, 1] + test_cm[1, 0])
test_specificity = test_cm[0, 0] / (test_cm[0, 0] + test_cm[0, 1])
```

```python
print(f'Test Accuracy: {test_acc:.4f}')
print(f'Test Error Rate: {test_err:.4f}')
print(f'Test F1 Score: {test_f1:.4f}')
print(f'Test Confusion Matrix:\n{test_cm}')
print("Test sensitivity: ", test_sensitivity)
print("Test specificity: ", test_specificity)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.svm import SVC

# Initialize logistic regression model
logistic_model = LogisticRegression()

# Fit logistic regression model
logistic_model.fit(x_train_smote, y_train_smote)

# Get probability estimates for logistic regression
logistic_probs = logistic_model.predict_proba(x_test)[:, 1]

# Calculate false positive rate (fpr) and true positive rate (tpr) for logistic regression
logistic_fpr, logistic_tpr, _ = roc_curve(y_test, logistic_probs)

# Calculate AUC score for logistic regression
logistic_auc = roc_auc_score(y_test, logistic_probs)

# Initialize SVM model
svm_model = SVC(probability=True)

# Fit SVM model
svm_model.fit(x_train_smote, y_train_smote)

# Get probability estimates for SVM
svm_probs = svm_model.predict_proba(x_test)[:, 1]

# Calculate false positive rate (fpr) and true positive rate (tpr) for SVM
svm_fpr, svm_tpr, _ = roc_curve(y_test, svm_probs)

# Calculate AUC score for SVM
svm_auc = roc_auc_score(y_test, svm_probs)

# Plot ROC curves for logistic regression and SVM
plt.figure(figsize=(8, 6))
plt.plot(logistic_fpr, logistic_tpr, linestyle='-', label=f'Logistic Regression (AUC = {logistic_auc:.2f})')
plt.plot(svm_fpr, svm_tpr, linestyle='-', label=f'SVM (AUC = {svm_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='orange', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()
```
